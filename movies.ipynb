```python
import ast
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):  # convert string into list
        list.append(i['name'])
    return list    
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
)
```

```python
def convert3(obj):
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list        
            
        
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head(3)
```

```python
def fetch_director(obj):
    list=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
    return list        
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# now we convert text i.e tags into vector to find the similarity btw the tags
# we use bag of words for textvectorization
# in bag of words technique---combine all tags i.e tags1+tags2.....tags5000 as a large text
# then we extract 5000  most common words used from large text
# now we find the frequency of all 5000 words from large text --w1|w2|w3....w5000.
#                                                             m1 5 3  0        1
#                                                             m3 1  6  3       0  (5000x5000)
# during vectoriztion we remove all stop words like are,a,the,in ....
```

```python
# now we apply stemming 
# ['loved','loving','loves'] then output willbe ['love','love','love']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    list=[]
    for i in text.split():
        list.append(ps.stem(i))
    return " ".join(list)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
for i in cv.get_feature_names_out() :
    print (i)
    
```

```python
# now we find cosine similarity --- actually find the angle btw two vectors
# angle = 1/distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# if sort the distances then we will lose the index
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
